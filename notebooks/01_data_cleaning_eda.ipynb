# 01 Data Cleaning and EDA

This notebook loads the Kaggle Global Weather Repository dataset, cleans it, and creates basic exploratory visualizations.

```python
import sys
from pathlib import Path

PROJECT_ROOT = Path.cwd().parents[0] if Path.cwd().name == 'notebooks' else Path.cwd()
sys.path.append(str(PROJECT_ROOT / 'src'))

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

from data_processing import load_weather_data, clean_weather_data, get_temperature_column, get_precipitation_column
from visualization import daily_average
```

```python
raw_path = PROJECT_ROOT / 'data' / 'raw' / 'GlobalWeatherRepository.csv'
df = load_weather_data(raw_path)
df = clean_weather_data(df)
df.head()
```

```python
df.info()
df.isna().sum().sort_values(ascending=False).head(20)
```

```python
temp_col = get_temperature_column(df)
daily_temp = daily_average(df, temp_col)

plt.figure(figsize=(12, 5))
plt.plot(daily_temp['date'], daily_temp[temp_col])
plt.title('Global Average Temperature Trend Over Time')
plt.xlabel('Date')
plt.ylabel('Average Temperature (C)')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```

```python
precip_col = get_precipitation_column(df)
if precip_col:
    daily_precip = daily_average(df, precip_col)
    plt.figure(figsize=(12, 5))
    plt.plot(daily_precip['date'], daily_precip[precip_col])
    plt.title('Global Average Precipitation Trend Over Time')
    plt.xlabel('Date')
    plt.ylabel('Average Precipitation')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
```

```python
plt.figure(figsize=(12, 8))
sns.heatmap(df.select_dtypes('number').corr(numeric_only=True), cmap='coolwarm', center=0)
plt.title('Correlation Heatmap')
plt.tight_layout()
plt.show()
```
